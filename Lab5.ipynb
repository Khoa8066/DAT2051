```python
import pandas as pd
```

# B1

```python
key1='product_id'
key2='product_name'
key3='price'
value_df1_1=[101,102,103]
value_df1_2=['Apple','Banana','Orange']
value_df1_3=[2.5,1.8,3.2]
value_df2_1=[104,105,106]
value_df2_2=['Mango','Grapes','Cherry']
value_df2_3=[4.0,2.7,3.5]
df1=pd.DataFrame({key1:value_df1_1,key2:value_df1_2,key3:value_df1_3})
df2=pd.DataFrame({key1:value_df2_1,key2:value_df2_2,key3:value_df2_3})
pd.concat((df1,df2),axis=0,ignore_index='False')
```

# B2

```python
left_df=pd.DataFrame(
    {'customer_id':[1,2,3],
     'product_id':[101,102,103],
     'quantity':[2,1,3]}
)
right_df=pd.DataFrame(
    {'product_id':[101,102,103],
     'price':[2.5,1.8,3.2]}
)

pd.merge(
    left=left_df,
    right=right_df,
    how='left',
    on='product_id'
)
```

# B3

```python
left_df=pd.DataFrame(
    {
        'key':['A','B','C'],
        'vl_left':[1,2,3]
    }
)
right_df=pd.DataFrame(
    {
        'key':['B','C','D'],
        'vl_right':[4,5,6]
    }
)
pd.merge(
    left=left_df,
    right=right_df,
    how='outer',
    on='key'
)
```

# B4

```python
left_df=pd.DataFrame(
    {
        'key':['A','B','C'],
        'vl_left':[1,2,3]
    }
)
right_df=pd.DataFrame(
    {
        'key':['B','C','D'],
        'vl_right':[4,5,6]
    }
)
third_df=pd.DataFrame(
    {
        'key':['A','C','E'],
        'vl_third':[7,8,9]
    }
)
pd.merge(
    left=pd.merge(
    left=left_df,
    right=right_df,
    how='outer',
    on='key'
),
    right=third_df,
    how='outer',
    on='key'
)
```
